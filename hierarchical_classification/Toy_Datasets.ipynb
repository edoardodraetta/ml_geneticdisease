```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# def gen_hierarchical_data():

n_families = 3
n_samples = 3

mean = [0, 0]
cov = [[1, 0], [0, 1]]
families = np.random.multivariate_normal(mean, cov, n_families)

X = np.zeros((n_families * n_samples, 2))
y = np.zeros((n_families * n_samples, n_families))

mean = [0, 0]
cov = [[0.1, 0], [0, 0.1]]
for i in range(n_families):
    X[i * n_samples : (i + 1) * n_samples, :] = families[i] + np.random.multivariate_normal(mean, cov, n_samples)
    y[i * n_samples : (i + 1) * n_samples, i] = 1


plt.figure(figsize=(5, 5))
plt.scatter(families[:, 0], families[:, 1], s=100, marker="x", label="sources")
for i in range(n_families):
    plt.scatter(X[i * n_samples : (i + 1) * n_samples, 0], X[i * n_samples : (i + 1) * n_samples, 1], marker=".")

lim = -3
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)
plt.show()
```

**What does it even mean to have hierarchical data?? This data is just three classes... right?**

```python
def gen_hierarchical_data(parents, n_parents, n_children, scale, make_plots=True, ax=None):

    if ax == None:
        ax = plt.gca()
    mean = [0, 0]
    cov = [[scale, 0], [0, scale]]

    X = np.zeros((n_parents * n_children, 2))
    y = np.zeros((n_parents*n_children, n_parents))
    
    for i in range(n_parents):
        noise = np.random.multivariate_normal(mean, cov, n_children)
        X[i * n_children : (i + 1) * n_children, :] = parents[i] + noise
        y[i * n_children : (i + 1) * n_children, i] = 1


    if make_plots:
        if n_parents == 1:
            ax.scatter(parents[0], parents[1], s=100, marker="x", label="sources")
            for i in range(n_parents):
                distrib_shape = plt.Circle(
                    (parents[0], parents[1]), 4 * scale, color="r", alpha=0.1, ec="black", fill=False
                )
                ax.add_patch(distrib_shape)
        else:
            ax.scatter(parents[:, 0], parents[:, 1], s=100, marker="x", label="sources")
            for i in range(n_parents):
                distrib_shape = plt.Circle(
                    (parents[i, 0], parents[i, 1]), 4 * scale, color="r", alpha=0.1, ec="black", fill=False
                )
                ax.add_patch(distrib_shape)

        for i in range(n_parents):
            ax.scatter(
                X[i * n_children : (i + 1) * n_children, 0], X[i * n_children : (i + 1) * n_children, 1], marker="."
            )

        lim = 3
        ax.set_xlim(-lim, lim)
        ax.set_ylim(-lim, lim)
        ax.title.set_text("{} parents, {} children".format(n_parents, n_children))
        plt.show()

    return X, y
```

```python
# np.random.seed(1)
level_0 = [0, 0]
level_1, labels_1 = gen_hierarchical_data(level_0, 1, 2, 1,)
level_2, labels_2 = gen_hierarchical_data(level_1, 2, 2, 0.2,)
level_3, labels_3 = gen_hierarchical_data(level_2, 4, 2, 0.1)
level_4, labels_4 = gen_hierarchical_data(level_3, 8, 2, 0.05)
level_5, labels_5 = gen_hierarchical_data(level_4, 16, 200, 0.01)
```
