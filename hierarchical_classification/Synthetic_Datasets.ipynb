```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
def gen_hierarchical_data(parents, n_parents, n_children, scale, make_plots=True, ax=None):

    if ax == None:
        ax = plt.gca()
    mean = [0, 0]
    cov = [[scale, 0], [0, scale]]

    X = np.zeros((n_parents * n_children, 2))
    y = np.zeros((n_parents * n_children, n_parents))

    for i in range(n_parents):
        noise = np.random.multivariate_normal(mean, cov, n_children)
        X[i * n_children : (i + 1) * n_children, :] = parents[i] + noise
        y[i * n_children : (i + 1) * n_children, i] = 1

    if make_plots:
        if n_parents == 1:
            ax.scatter(parents[0], parents[1], s=100, marker="x", label="sources")
            for i in range(n_parents):
                distrib_shape = plt.Circle(
                    (parents[0], parents[1]), 4 * scale, color="r", alpha=0.1, ec="black", fill=False
                )
                # ax.add_patch(distrib_shape)
        else:
            ax.scatter(parents[:, 0], parents[:, 1], s=100, marker="x", label="sources")
            for i in range(n_parents):
                distrib_shape = plt.Circle(
                    (parents[i, 0], parents[i, 1]), 4 * scale, color="r", alpha=0.1, ec="black", fill=False
                )
                # ax.add_patch(distrib_shape)

        for i in range(n_parents):
            ax.scatter(
                X[i * n_children : (i + 1) * n_children, 0], X[i * n_children : (i + 1) * n_children, 1], marker="."
            )

        lim = 3
        ax.set_xlim(-lim, lim)
        ax.set_ylim(-lim, lim)
        ax.title.set_text("{} parents, {} children".format(n_parents, n_children))
        plt.show()

    return X, y
```

```python
# np.random.seed(1)
level_0 = [0, 0]
level_1, labels_1 = gen_hierarchical_data(level_0, 1, 2, 1, make_plots=True)
level_2, labels_2 = gen_hierarchical_data(level_1, 2, 2, 0.2, make_plots=True)
level_3, labels_3 = gen_hierarchical_data(level_2, 4, 2, 0.1, make_plots=False)
level_4, labels_4 = gen_hierarchical_data(level_3, 8, 2, 0.05, make_plots=False)
level_5, labels_5 = gen_hierarchical_data(level_4, 16, 200, 0.01, make_plots=True)
```

## ML

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_predict, cross_val_score, train_test_split

n_ancestors = 3
n_families = 2  # families per ancestor
n_points = 300  # points per family
n_dims = 2  # dimensions per point

noise = np.ones(shape=(n_dims, 1)) * np.logspace(0, -1, num=3)

X = np.zeros(shape=(n_points * n_families, n_dims))
Y = np.zeros(shape=(n_points * n_families, 1))

N = n_points * n_families
ancestors = np.random.randn(n_ancestors, n_dims)
families = np.random.randn(n_families * n_ancestors, n_dims)

for i in range(N):

    j = int(i * n_ancestors / n_families / n_points)  # ancestor index
    l = int(i * n_ancestors / n_points) - 2 * j  # descendant index
    k = int(i * n_ancestors / n_points)
    Y[i] = k
    X[i, :] = noise[:, 0] * ancestors[j, :] + noise[:, 1] * families[l, :] + noise[:, 2] * np.random.randn(1, n_dims)


# Put into labeled dataframe
label_names = ["y"]
feature_names = []
for i in range(n_dims):
    feature_names.append("x" + str(i + 1))

df = pd.DataFrame(np.c_[Y, X], columns=label_names + feature_names)
sns.scatterplot(data=df, x="x1", y="x2", hue="y")
# plt.gca().set_aspect('equal')
plt.show()

print("Noise scaling is:", noise[0, :])

X_train, X_test, y_train, y_test = train_test_split(X, Y.ravel(), test_size=0.2, random_state=42)

# model
print("\nUsing random forest.\n")
ran_for = RandomForestClassifier(n_estimators=100, random_state=42)  
ran_for.fit(X_train, y_train)

# scores
cv_pred = cross_val_predict(ran_for, X_train, y_train, cv=3)
pred = ran_for.predict(X_test)
print("Accuracy:\n", accuracy_score(y_test, pred))
print("Confusion matrix:\n", confusion_matrix(y_train, cv_pred))
```
