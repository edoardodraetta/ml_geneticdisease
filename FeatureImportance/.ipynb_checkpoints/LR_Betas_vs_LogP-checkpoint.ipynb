```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

pd.set_option("display.max_rows", 85)
sns.set_theme(context="paper", font_scale=1.5, style="ticks", rc={"axes.grid": True})
pd.set_option('display.float_format', lambda x: '%.7f' % x)
```

# Plot Feature Betas vs -Log(p value)

```python
# Load beta and p-values from PP and sort alphabetically by feature name
df = pd.read_csv("../data/write.abnormal_allpcs_summary.csv", index_col=0, usecols=[1,2,3])
df.sort_index(inplace=True)

# Load my betas from logistic regression and sort by name
my_df = pd.read_csv("./data/lr/lr_betas.csv", index_col=0)         
my_df.index = my_df.index.str.replace(' ', '')
my_df.sort_index(inplace=True)
```

```python
# Check that everything is compatible
print(df.index.values)
print(my_df.index.values)
print()
print("Are the index names identical?")
df.index == my_df.index
```

```python
# Place into a comparison dataframe
df["my_beta"] = my_df.Beta.values 
df["beta_diff"] = abs(df["beta"].values) - abs(df["my_beta"].values)
df["beta_perc_diff"] = (df["beta_diff"].values / df["beta"].values) * 100
```

$$ BPD = \frac{|\beta_{pp}| - |\beta_{ed}|}{\beta_{pp}} $$

## Plot of percent difference

```python
sns.lineplot(x=np.arange(82), y=df["beta_perc_diff"])
plt.title("Percent difference between my beta and yours")
plt.xlabel("Feature Order in Dataframe")
plt.show()
```
